#### Activity 1: Operator Overloading for `Rectangle`

Create a class `Rectangle` having two private instance variables `length` and `breadth`. Add getter and setter functions for both the private variables. Create a function to calculate the area of the rectangle:
$$\text{area of rectangle = length * breadth}$$

Create two `Rectangle` objects `rect1` and `rect2` and add both the objects by overloading the `+` operator as follows:
$$rect3 = rect1 + rect2$$

Adding the two `Rectangle` objects should give another `Rectangle` object `rect3` whose length is the sum of the lengths of the two `Rectangle` objects and whose breadth is the sum of the breadths of the two `Rectangle` objects.


Also, calculate the area of the `rect3` Rectangle object.

**Hint:** Overload the magic function `__add__()`




```
#  Create a class 'Rectangle'. Add constructor to initialise variables
class Rectangle :

    def __init__(self,length,breadth):
        self.__length = length
        self.__breadth = breadth

    def get_length(self):
        return self.__length

    def get_breadth(self):
        return self.__breadth

    def set_length(self,new_length):
        self.__lenght = new_length

    def set_breadth(self,new_breadth):
        self.__breatdh = new_breadth

    def __add__(self,other):
        new_length = self.__length + other.__length
        new_breadth = self.__breadth + other.__breadth
        new = Rectangle(new_length,new_breadth)
        return new

    def get_area(self):
        return self.__length * self.__breadth

rect1 = Rectangle(4,6)
rect2 = Rectangle(5,10)
rect3 = rect1 + rect2
print('Length of rect1 =',rect1.get_length())
print('Length of rect2 =',rect2.get_length())
print('Length of rect1 + length of rect2 =',rect1.get_length() + rect2.get_length())
print('Equal to length of rect3 =',rect3.get_length())
print('-' * 50)
print('Breadth of rect1 =',rect1.get_breadth())
print('Breadth of rect2 =',rect2.get_breadth())
print('Breadth of rect1 + breadth of rect2 =',rect1.get_breadth() + rect2.get_breadth())
print('Equal to breadth of rect3 =',rect3.get_breadth())
print('-' * 50)
print('Area of rect1 in square unit =',rect1.get_area())
print('Area of rect2 in square unit =',rect2.get_area())
print('Area of rect3 in square unit =',rect3.get_area())
```
